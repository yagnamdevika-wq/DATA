```python
import pandas as pd
```

```python
df=pd.read_csv("/content/bank_customer_data.csv")
df
```

```python
df["customer_name"] = df["customer_name"].str.title()
df
```

```python
df["email"]=df["email"].str.lower()
df
```

```python
df["bank_name"]=df["bank_name"].str.upper()
df
```

```python
df["account_type"]=df["account_type"].str.upper()
df
```

```python
df["balance"]=df["balance"].str.replace("Rs", "",regex=False)
df
```

```python
df["balance"]=df["balance"].str.replace("$", "",regex=False)
df
```

```python
df["balance"]=df["balance"].str.replace(",", "",regex=False)
df
```

```python
df["balance"] = df["balance"].astype(int)
df
```

```python
df["username"] = df["email"].str.split("@").str[0]
df
```

```python
df["domain"] = df["email"].str.split("@").str[1]
df
```

```python
df["branch_code"] = df["branch_city"].str[:3].str.upper()
df
```

```python
df["account_number"] = df["account_number"].str.replace("ACC-", "", regex=False)
df
```
